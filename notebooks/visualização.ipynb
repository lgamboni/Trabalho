```python
trace = go.Scatter(
    x = bch.date,
    y = bch.close
)

layout = go.Layout(
    width = 800,
    height = 500,
     title = 'Cotação BCD/BTG  - close'
)
data = [trace]
fig = go.Figure(data=data, layout=layout)

iplot(fig,filename='basic-line',show_link=False)
```

Por alguma razão, a exchange dagatecoin não está fornecendo os valores low. Logo, não é possível fazer o gráfico **Candlestick** que se trata de um tipo de gráfico que mostra se o preço da moeda está subindo ou descendo em relação ao valor, imediatamente, anteiror.

```python
trace = go.Scatter(
    x = ltc.date,
    y = ltc.volume,
    name = 'LTC/BTC'
)
trace1 = go.Scatter(
    x = bch.date,
    y = bch.volume,
    name = 'BCH/BTC'
)
trace2 = go.Scatter(
    x = eth.date,
    y = eth.volume,
    name = 'ETH/BTC'
)


layout = go.Layout(
    width = 800,
    height = 500,
     title = 'Cotação BCD/BTG e BTG/BTC e ETH/BTC  - close'
)
data = [trace,trace1,trace2]
fig = go.Figure(data=data, layout=layout)

iplot(fig,filename='basic-line',show_link=False)
```
